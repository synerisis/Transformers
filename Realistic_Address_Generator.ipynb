```python
import random
import pandas as pd

def tumble_address(address):
    """Applies various "tumbling" techniques to an address."""
    tumbled_address = address

    # 1. Adding random, non-existent unit numbers or apartment indicators.
    units = [" Apt " + str(random.randint(100, 999)), " Unit " + str(random.randint(1, 50)), " # " + str(random.randint(1, 200))]
    if random.random() > 0.5:
        tumbled_address += random.choice(units)

    # 2. Randomly changing or inserting characters within the street name or city.
    address_parts = tumbled_address.split(',')
    if len(address_parts) > 1:
        street_city = address_parts[0].strip()
        if random.random() > 0.3:
            index_to_change = random.randint(0, len(street_city) - 1)
            tumbled_address = tumbled_address[:index_to_change] + random.choice('abcdefghijklmnopqrstuvwxyz0123456789') + tumbled_address[index_to_change + 1:]
        if random.random() > 0.3:
            index_to_insert = random.randint(0, len(street_city))
            tumbled_address = tumbled_address[:index_to_insert] + random.choice('abcdefghijklmnopqrstuvwxyz0123456789') + tumbled_address[index_to_insert:]

    # 3. Replacing the correct street type with an incorrect or a less common one.
    street_types = {"Street": ["Strt", "Strete"], "Avenue": ["Avnue", "Avenoo"], "Road": ["Rd.", "Rode"], "Boulevard": ["Blvd.", "Boulvard"]}
    for correct, incorrect_list in street_types.items():
        if correct in tumbled_address:
            if random.random() > 0.4:
                tumbled_address = tumbled_address.replace(correct, random.choice(incorrect_list))
            break # Assume only one street type per address

    # 4. Introducing common misspellings to parts of the address.
    misspellings = {"California": "Calfornia", "New York": "Nw York", "Texas": "Texaas", "Florida": "Floridda", "Washington": "Washignton"}
    for correct, misspelled in misspellings.items():
        if correct in tumbled_address:
            if random.random() > 0.5:
                tumbled_address = tumbled_address.replace(correct, misspelled)
            break

    return tumbled_address

real_addresses = [
    "1600 Pennsylvania Ave NW, Washington, DC 20500",
    "221B Baker St, London, NW1 6XE", # Including a non-US address to show robustness
    "Statue of Liberty, New York, NY 10004",
    "Golden Gate Bridge, San Francisco, CA 94129",
    "White House, 1600 Pennsylvania Ave NW, Washington, DC 20500",
    "Empire State Building, 20 W 34th St, New York, NY 10001",
    "Hollywood Sign, Los Angeles, CA 90068",
    "Space Needle, Seattle, WA 98109",
    "Mount Rushmore, Keystone, SD 57751",
    "Grand Canyon National Park, AZ 86023",
    "Yosemite National Park, CA 95389",
    "Yellowstone National Park, WY 82190",
    "Walt Disney World, Orlando, FL 32830",
    "Las Vegas Strip, Las Vegas, NV 89109",
    "French Quarter, New Orleans, LA 70116",
    "Navy Pier, Chicago, IL 60611",
    "Independence Hall, Philadelphia, PA 19106",
    "The Alamo, San Antonio, TX 78205",
    "Pike Place Market, Seattle, WA 98101",
    "Fisherman's Wharf, San Francisco, CA 94133",
    "Miami Beach, FL 33139",
    "Boston Common, Boston, MA 02116",
    "Gateway Arch, St. Louis, MO 63102",
    "Niagara Falls, NY 14303",
    "Great Smoky Mountains National Park, TN 37738",
    "Rocky Mountain National Park, CO 80517",
    "Zion National Park, UT 84767",
    "Arches National Park, UT 84532",
    "Grand Teton National Park, WY 83012",
    "Death Valley National Park, CA 92328",
    "Everglades National Park, FL 33332",
    "Denali National Park and Preserve, AK 99755",
    "Acadia National Park, ME 04609",
    "Great Sand Dunes National Park and Preserve, CO 81131",
    "Bryce Canyon National Park, UT 84764",
    "Canyonlands National Park, UT 84532",
    "Capitol Reef National Park, UT 84775",
    "Mesa Verde National Park, CO 81330",
    "Petrified Forest National Park, AZ 86028",
    "Saguaro National Park, AZ 85743",
    "Shenandoah National Park, VA 22835",
    "Virgin Islands National Park, VI 00830",
    "Voyageurs National Park, MN 56627",
    "Wind Cave National Park, SD 57567",
    "Wrangell-St. Elias National Park and Preserve, AK 99566",
    "Abraham Lincoln Birthplace National Historical Park, KY 42748",
    "Frederick Douglass National Historic Site, DC 20020",
    "Martin Luther King, Jr. National Historical Park, GA 30312",
    "Rosie the Riveter/World War II Home Front National Historical Park, CA 94801",
    "Manzanar National Historic Site, CA 93545",
]
# 3. Apply the tumbling function to each of the 50 real addresses
tumbled_addresses = [tumble_address(address) for address in real_addresses]

# 4. Store both the original and the tumbled addresses in a suitable data structure
df_addresses = pd.DataFrame({'original_address': real_addresses, 'tumbled_address': tumbled_addresses})

display(df_addresses)
```
